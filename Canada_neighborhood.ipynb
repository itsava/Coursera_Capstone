```python
import numpy as np
import urllib.request,urllib.parse,urllib.error
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import json 
import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
```

# Parsing HTML link using BeautifulSoup

```python
url=input('Enter the url')
html=urllib.request.urlopen(url).read()
soup=BeautifulSoup(html,'html.parser')
p = soup.find_all('tr')
para =[]
real=[]
for x in p:
    para.append(str(x))
for n in para:
    ss=n.split("\n")
    ss[1]=ss[1][4:]
    ss[3]=ss[3][4:]
    ss[5]=ss[5][4:]
    del ss[0]
    del ss[1]
    del ss[2]
    del ss[3]
    if ss[1]!='Not assigned':
        real.append(ss)
del real[104:]
hea=real[0]
del real[0]
```

# Creating and Modifying Dataframe 

```python
df=pd.DataFrame(real,columns=hea)
df2=df.groupby(["Postal Code","Borough"]).agg({"Neighborhood":','.join})
df2
```

# Finding Coordinates of each postal code

```python
url="http://cocl.us/Geospatial_data"
df3=pd.read_csv(url,index_col=False )
df4=pd.merge(df,df3,left_on="Postal Code",right_on="Postal Code",how="right")
df4
```

# For the Toronto neighborhood

```python
torronto_data = df4[df4['Borough'] == 'Downtown Toronto'].reset_index(drop=True)
torronto_data1 = df4[df4['Borough'] == 'East Toronto'].reset_index(drop=True)
torronto_data2 = df4[df4['Borough'] == 'Central Toronto'].reset_index(drop=True)
torronto_data3 = df4[df4['Borough'] == 'West Toronto'].reset_index(drop=True)
torronto_data=torronto_data.append(torronto_data1)
torronto_data=torronto_data.append(torronto_data2)
torronto_data=torronto_data.append(torronto_data3)
torronto_data
```

```python
address = 'Toronto, Canada'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
```

# Creating the map of neighborhoods of Toronto

```python
map_torronto = folium.Map(location=[latitude, longitude], zoom_start=11)
# add markers to map
for lat, lng, label in zip(torronto_data['Latitude'], torronto_data['Longitude'], torronto_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat, lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_torronto)
map_torronto
```
